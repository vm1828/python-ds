```python
import tensorflow as tf
```

# Shape of input/output of Embedding and Flatten layers

```python
# input parameters
vocab_size = 10000
embedding_dim = 100
max_length = 100
batch_size = 32

# model
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_length),
    tf.keras.layers.Flatten()
])

# display input and output shapes for each layer without data (axis 0 will be None)
for layer in model.layers:
    input_shape = layer.input_shape
    output_shape = layer.output_shape
    print(f"Layer: {layer.name}\n\tInput shape: {input_shape}\n\tOutput shape: {output_shape}")
print("="*30)

# generate input data
input_data = tf.random.uniform(shape=(batch_size, max_length), minval=0, maxval=vocab_size, dtype=tf.int32)

# put data to each layer manually and display shapes
embedding_layer_output = model.layers[0](input_data)
flatten_layer_output = flattened_output = model.layers[1](embedding_layer_output)
# flattened_output = tf.reshape(embedding_layer_output, shape=(batch_size, -1))

print("\nData flow:")
print(f"Embedding: {input_data.shape} --> {embedding_layer_output.shape}")
print(f"Flatten: {embedding_layer_output.shape} --> {flatten_layer_output.shape}")
```
