# Line Plots

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

linear_data = np.array([1,2,3,4,5,6,7,8])
exponential_data = linear_data**2

plt.figure()
# plot the linear data and the exponential data
plt.plot(linear_data, '-o', exponential_data, '-o')
# we only give y-axis values to our plot call
# plot function uses index of the series as the x value

# plot another series with a dashed red line
plt.plot([22,44,55], '--r')

plt.xlabel('Some data')
plt.ylabel('Some other data')
plt.title('A title')
# add a legend with legend entries (because we didn't have labels when we plotted the data series)
plt.legend(['Baseline', 'Competition', 'Us'])

# fill the area between the linear data and exponential data
plt.gca().fill_between(range(len(linear_data)), 
                       linear_data, exponential_data, 
                       facecolor='blue', 
                       alpha=0.25) # alpha - transparency value
```

Let's try working with dates

```python
plt.figure()

observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')

plt.xticks(rotation=90)
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o')
```

Earlier versions (?) of matplotlib expected standard library `datetime.date` data type and didn't support numpy dates. In order to covert numpy dates into `datetime.date` pandas can be used

```python
import pandas as pd

plt.figure()
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates)) 
# trying to plot a map will result in an error bcz map function returns an iterator
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o')

# we can rotate tick labels for x-axis because they are overlapping 
x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
# adjust the subplot so the text doesn't run off the image
# plt.subplots_adjust(bottom=0.25)

ax = plt.gca()
ax.set_xlabel('Date')
ax.set_ylabel('Units')
ax.set_title('Exponential vs. Linear performance')
# you can add mathematical expressions in any text element
ax.set_title("Exponential ($x^2$) vs. Linear ($x$) performance")
```

Matplotlib has a fairly strong connection to LaTeX - type setting language used by mathematicians and scientists (https://en.wikipedia.org/wiki/LaTeX)
This means that you can use a subset of LaTeX directly in your labels, then matplotlib will render them as equations.
