# Heatmaps

Heatmaps are a way to visualize 3-dimensional data and to take advantage of spatial proximity of those dimensions.  
  
Great candidate for heatmap is weather data: we have 2 dimensions - latitude and longitude, and then we can overlay on top of this a third dimension - temperature, rainfall amounts etc., and use color to indicate its intensity.  

An interesting example is a heatmap showing the probability of a place where the plane has crashed, where more red color indicates higher probability and blue color - lower probability.

In matplotlib a heatmap is simply a 2-dimansional histogram where the x and the y values indicate potential points and the color plotted is the frequency of the observation.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.figure()

Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)
_ = plt.hist2d(X, Y, bins=25)
```

```python
# increase the number of bins
_ = plt.hist2d(X, Y, bins=100)

# add a colorbar legend
plt.colorbar()
# under the hood pyplot is finding the current rendered image, 
# generating the colorbar, resizing the existing axes and using gridspec 
# to add a new axes the size of the colorbar.
```
